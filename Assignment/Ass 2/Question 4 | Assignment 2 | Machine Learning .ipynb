```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
np.random.seed(12)
```

```python
x = np.linspace(0, 1, 50)
c = 1
m = 1
noise = 0.3*np.random.rand(50)
y = m * x + c + noise
p = np.poly1d(np.polyfit(x, y, 5))
```

```python
t = np.linspace(0, 1, 200)
plt.plot(x, y, 'o', t, p(t), '-')
plt.show()
```

```python
x = np.linspace(0, 1, 50)
c = 1
m = 1
noise = 0.3*np.random.rand(50)
y = m * x + c + noise
avg = []

for i in range (100) :
    xr = []
    yr = []
    for j in range (30) :
        r = np.random.randint (49)
        xr.append (x[r])
        yr.append (y[r])
    p = np.poly1d (np.polyfit(xr, yr, 5))
    avg += [p.c] 

av=np.zeros(6)
for i in range(100):
    for j in range(6):
        av[j]+=avg[i][j]

av=av/100
print (av)

av = np.lib.polynomial.poly1d (av)
t = np.linspace(0, 1, 200)
plt.plot(x, y, 'o', t, av(t), '-')
plt.show()
```
