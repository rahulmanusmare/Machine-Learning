#### (a) Write a function normalEquationRegression(X, y) where X is our feature matrix containing N samples (rows) and d features (columns) and y is our output vector containing N samples. This function returns a vector θ containing d + 1 rows. You are free to use numpy’s matrix inverse, determinant and multiplication routines.

```python
X = np.matrix ('1; 3; 6')
y = np.matrix ('6; 10; 16')
```

```python
def normalEquationRegression (X, y) :
    X = np.concatenate ((np.ones ((np.size (X, 0), 1), dtype = int), X), axis = 1)
    return np.matmul (inv (np.matmul(X.transpose (),X)),np.matmul (X.transpose (),y))
print (normalEquationRegression (X, y))
```

#### (b) Write a function gradientDescentRegression (X, y, alpha = 0.1) to learn the regression coefficients using gradient descent. You have to write the formulae for gradient wrt the different θj∀j ∈ (1, ..d)

```python
def gradientDescentRegression(X, y, alpha = 0.01):
    X = X.T
    N,d = X.shape [0],X.shape [1]
    b = np.ones ((N,1))
    X = np.hstack ((b,X))
    theta = [0] * (d + 1)
    for itr in range (5):
        error = np.matmul (X,theta)- y
        theta = theta - (2*alpha / N) * np.matmul (X.T, error)
    
    return theta

x = np.array ([[1,2,3,4],[2,4,6,8]])
y = np.array ([2,3,4,5])
print (gradientDescentRegression (x,y))
```

#### (c) Write a function gradientDescentAutogradRegression(X, y, alpha = 0.1) to learn the regression coeffi-cients using gradient descent. Instead of writing the formulae for computing gradients by yourself, you will use Autograd to automatically do that for you.

```python
def gradientDescentAutogradRegression (X, y, alpha = 0.01):
    
    def cost (theta):
        e = np.dot(X,theta) - y
        return np.dot(e.T, e)
    
    X = X.T
    N,d = X.shape [0], X.shape [1]
    b = np.ones ((N,1))
    X = np.hstack ((b,X))
    theta = np.zeros (d+1)
    
    for itr in range (1000):
        grad_err = grad (cost)
        delta = grad_err(theta)
        theta = theta - alpha * delta / N
    return theta
    
x = np.array([[1,2,3,4],[2,4,6,8]])
y = np.array([2,3,4,5])

print(gradientDescentAutogradRegression(x,y))
```
