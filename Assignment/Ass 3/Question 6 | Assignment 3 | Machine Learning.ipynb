#### (a) Create a contour plot in the θ0 and θ1 space of the residual sum of squares

```python
def compute_error(b,m,points):
    totalError = 0 
    for i in range (0, len(points)):
        x = points[i][0]
        y = points[i][1]
        totalError += (y-(m*x + b)) ** 2
    return totalError

points = [[1,6],[3,10],[6,16]]
X = np.arange(-100,100,1)
Y = np.arange(-100,100,1)

Z=[[0]*len(X) for i in range(len(Y))]

for i in range(len(X)):
    for j in range(len(Y)):
        Z[i][j]=compute_error(X[i],Y[j],points)
        
        
cont=plt.contour(X, Y, Z)
plt.clabel(cont,inline=1,fontsize=8)

plt.show()
```
