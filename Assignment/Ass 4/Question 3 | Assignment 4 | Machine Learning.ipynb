```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from matplotlib.animation import FuncAnimation
from IPython.display import HTML, Image


%matplotlib inline

plt.style.use('seaborn-white')


import numpy as np
import matplotlib.pyplot as plt



def err(b,m,x_c,y_c):
    totalError = 0 
    for i in range (0, len(x_c)):
        y=y_c[i]
        x=x_c[i]
        totalError += (y-(m*x + b)) ** 2
    return totalError**0.5

x=[1,3,6]
y=[6,10,16]
X = np.arange(-20,20,0.1)
Y = np.arange(-20,20,0.1)

Z=[]
for i in range(len(Y)):
    Z+=[[]]
for j in range(len(X)):
    Z[-1]+=[0]


for i in range(len(X)):
    for j in range(len(Y)):
        Z[i][j]=err(X[i],Y[j],x,y)
       
       
#cont=plt.contour(X, Y, Z)
#plt.clabel(cont,inline=2,fontsize=10)
fig1, ax1 = plt.subplots(figsize = (7,7))
ax1.contour(X, Y, Z)

w_s=[]
b_s=[]
# Create animation
line, = ax1.plot([], [], 'r', label = 'Gradient descent', lw = 1.5)
point, = ax1.plot([], [], '*', color = 'red', markersize = 4)
value_display = ax1.text(8, 6, '', transform=ax1.transAxes)

w_s=[4]+w_s
b_s=[2]+b_s

def init_1():
    line.set_data([], [])
    point.set_data([], [])
    value_display.set_text('')

    return line, point, value_display

def animate_1(i):
    # Animate line
    line.set_data(b_s[:i], w_s[:i])

    #print (b_s[i])
    # Animate points
    point.set_data(b_s[i], w_s[i])

    # Animate value display
    #value_display.set_text('Min = ' + str(J_history_reg[i]))

    return line, point, value_display

ax1.legend(loc = 1)

anim1 = FuncAnimation(fig1, animate_1, init_func=init_1,
                              frames=len(b_s), interval=100, 
                              repeat_delay=60, blit=True)
#plt.show()
#anim1.save('contour_prior.gif', dpi=80, writer='pillow')
HTML(anim1.to_jshtml())
```
